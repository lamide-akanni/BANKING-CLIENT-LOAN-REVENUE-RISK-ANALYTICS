```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv(r'C:\Users\lamid\Desktop\Banking\Banking.csv')

# First 5 rows
df.head(5)
```

```python
# Parse the date column correctly
df['Joined Bank'] = pd.to_datetime(df['Joined Bank'], format='%d-%m-%Y')
print(df['Joined Bank'].dtype)
```

**Create the Income Band Column**


```python
# Define the bin boundaries
bins = [0, 100000, 300000, float('inf')]
labels = ['Low', 'Mid', 'High']

# Create the Income Band column
df['Income Band'] = pd.cut(
    df['Estimated Income'],
    bins=bins,
    labels=labels,
    include_lowest=True
)

# Check the distribution
print(df['Income Band'].value_counts())
df['Income Band'].value_counts().plot(kind='bar')
plt.show()
```

**Categorical Analysis; Value Counts & Countplots**

```python
# Define categorical columns to analyse
categorical_cols = [
    "BRId", "GenderId", "IAId", "Amount of Credit Cards",
    "Nationality", "Occupation", "Fee Structure",
    "Loyalty Classification", "Properties Owned",
    "Risk Weighting", "Income Band"
]

# Print value counts for each
for col in categorical_cols:
    print(f"\nValue Counts for '{col}':")
    print(df[col].value_counts())

# Univariate countplots — one chart per column
for i, col in enumerate(categorical_cols):
    plt.figure(i)
    sns.countplot(data=df, x=col)
    plt.title(f'Distribution of {col}')
    plt.show()

# Bivariate — break down each category by Nationality
for i, col in enumerate(categorical_cols):
    plt.figure(i)
    sns.countplot(data=df, x=col, hue='Nationality')
    plt.show()
```


**Numerical Analysis; Histograms with KDE**

```python
numerical_cols = [
    'Estimated Income', 'Superannuation Savings',
    'Credit Card Balance', 'Bank Loans', 'Bank Deposits',
    'Checking Accounts', 'Saving Accounts',
    'Foreign Currency Account', 'Business Lending'
]

# Plot all in a grid — 3 rows × 3 columns
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(3, 3, i+1)
    sns.histplot(df[col], kde=True)
    plt.title(col)
plt.tight_layout()
plt.show()
```

**Correlation Heatmap**

A correlation matrix to see which financial variables move together

```python
# Compute correlations between all numerical columns
correlation_matrix = df[numerical_cols].corr()

# Draw the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap='coolwarm',
    fmt=".2f"
)
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

Bank Deposits has the strongest positive correlation with Checking Accounts, Saving Accounts, and Foreign Currency Account; meaning clients with high deposits tend to hold funds across multiple account types.

**Regression Plots for Key Variable Pairs**

 plots to visually confirm the relationships spotted in the heatmap.

```python
pairs_to_plot = [
    ('Bank Deposits', 'Saving Accounts'),
    ('Checking Accounts', 'Saving Accounts'),
    ('Checking Accounts', 'Foreign Currency Account'),
    ('Age', 'Superannuation Savings'),
    ('Estimated Income', 'Checking Accounts'),
    ('Bank Loans', 'Credit Card Balance'),
    ('Business Lending', 'Bank Loans'),
]

# Create the grid — 3 rows × 3 columns = 9 slots (7 used, 2 empty)
fig, axes = plt.subplots(3, 3, figsize=(18, 15))

# Flatten the grid so we can loop through it easily
axes = axes.flatten()

# Loop through each pair and plot in the next available slot
for i, (x_col, y_col) in enumerate(pairs_to_plot):
    sns.regplot(
        data=df,
        x=x_col,
        y=y_col,
        ax=axes[i],                        # place in correct grid slot
        scatter_kws={'alpha': 0.4},
        line_kws={'color': 'red'}
    )
    axes[i].set_title(f'{x_col} vs {y_col}', fontsize=10)

# Hide the 2 empty slots (position 8 and 9)
for j in range(len(pairs_to_plot), len(axes)):
    axes[j].set_visible(False)

plt.suptitle('Regression Plots — Variable Relationships', 
             fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```
